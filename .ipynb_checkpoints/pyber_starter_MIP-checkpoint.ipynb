```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.patches as mpatches

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)


#group by to get the count of the rides and the average of the fare
ride_df = (ride_data.groupby('city', as_index=False)
        .agg({'ride_id':'count', 'fare':'mean', })
        .rename(columns={'ride_id':'ride_count','fare':'average_fare'}))

# Combine the data into a single dataset
merge_table = pd.merge(ride_df, city_data, on="city", how="outer")


#Display the data table for preview
merge_table.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

urban_df = merge_table.loc[merge_table["type"] == "Urban",:]
urban_x = list(urban_df["ride_count"])
urban_y = list(urban_df["average_fare"])
urban_z = list(urban_df["driver_count"]*10)

sub_df = merge_table.loc[merge_table["type"] == "Suburban",:]
sub_x = list(sub_df["ride_count"])
sub_y = list(sub_df["average_fare"])
sub_z = list(sub_df["driver_count"]*10)

rural_df = merge_table.loc[merge_table["type"] == "Rural",:]
rural_x = list(rural_df["ride_count"])
rural_y = list(rural_df["average_fare"])
rural_z = list(rural_df["driver_count"]*10)


# Build the scatter plots for each city types
plt.scatter(urban_x, urban_y, s=urban_z, c="#F08080", alpha = 0.6, edgecolors="black", linewidth=1, label = "Urban")
plt.scatter(sub_x, sub_y, s=sub_z, c="#85C1E9", alpha = 0.6, edgecolors="black", linewidth=1, label = "Suburban")
plt.scatter(rural_x, rural_y, s=rural_z, c="#F8C471", alpha = 0.6, edgecolors="black", linewidth=1, label = "Rural")

# Incorporate the other graph properties
plt.xlabel("Number of Rides")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")
plt.grid(True)

# Create a legend
legend = plt.legend(title = "City Types")
for key in legend.legendHandles:
    key.set_sizes([40])


# Incorporate a text label regarding circle size
plt.text(42,37,"Note: Circle Size correlates to number of drivers")

# Show Figure
plt.show()

# Show Figure
plt.savefig("Pyber Ride Sharing Data (2016).png")
```

## Total Fares by City Type

```python
# Calculate Type Percents
#create a new merge table that doesnt average the fare 
merge_table2 = pd.merge(ride_data, city_data, on="city", how="outer")
fare_subset = merge_table2.groupby(['type'])
fare_summary = pd.DataFrame(fare_subset["fare"].sum())

# Build Pie Chart
colors = [ "#F8C471", "#85C1E9","#F08080"]
explode = (0, 0, 0.1)
fare_summary.plot(kind = "pie",y=fare_summary.keys(), explode=explode,colors=colors,
          autopct="%1.1f%%",startangle=150, title ="% of Total Fares by City Type", legend= False,shadow=True)
plt.axis('off')
# Show Figure
plt.show()

# Save Figure
plt.savefig("Total Fares by City Type.png")
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rides_subset = merge_table2.groupby(['type'])
rides_summary = pd.DataFrame(rides_subset["ride_id"].nunique())

# Build Pie Chart
colors = [ "#F8C471", "#85C1E9","#F08080"]
explode = (0, 0, 0.1)
rides_summary.plot(kind = "pie",y=rides_summary.keys(), explode=explode,colors=colors,
          autopct="%1.1f%%",startangle=150, title ="% of Total Rides by City Type", legend= False,shadow=True)
plt.axis('off')

# Show Figure
plt.show()

# Save Figure
plt.savefig("Total Rides by City Type.png")
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
driver_subset = city_data.groupby(['type'])
driver_summary = pd.DataFrame(driver_subset["driver_count"].sum())

# Build Pie Chart
colors = [ "#F8C471", "#85C1E9","#F08080"]
explode = (0, 0, 0.1)
driver_summary.plot(kind = "pie",y=driver_summary.keys(), explode=explode,colors=colors,
          autopct="%1.1f%%",startangle=200, title ="% of Total Drivers by City Type", legend= False,shadow=True)
plt.axis('off')

# Show Figure
plt.show()

# Save Figure
plt.savefig("Total Drivers by City Type.png")
```
